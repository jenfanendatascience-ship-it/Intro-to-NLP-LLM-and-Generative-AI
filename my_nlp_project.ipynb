## Objective

This project demonstrates a simple autocomplete system using a Bigram Language Model.

The system learns which words usually follow one another and predicts the most likely next word.

## NLP Concepts Used

- Tokenization
- Dictionaries
- Word Frequency
- Bigram Language Model
- Next Word Prediction
- Text Generation

This project was developed using Python and inspired by the concepts covered in the NELIREF NLP course.

```python
from collections import Counter
import string
```

```python
def tokenize(text):
    """Split text into clean lowercase tokens."""

    words = text.split()

    tokens = []

    for word in words:
        word = word.lower()
        word = word.strip(string.punctuation)

        if word:
            tokens.append(word)

    return tokens
```

```python
corpus = """
the cat sat on the mat
the dog sat on the floor
the cat ran to the door
the dog chased the cat
the cat climbed the tree
the dog barked loudly
"""

tokens = tokenize(corpus)

print(tokens)
```

```python
def build_bigram_model(tokens):
    """Build a dictionary where each word maps to the words that followed it."""

    model = {}

    for i in range(len(tokens) - 1):

        current_word = tokens[i]
        next_word = tokens[i + 1]

        if current_word not in model:
            model[current_word] = []

        model[current_word].append(next_word)

    return model
```

```python
def predict_next_word(word, model):
    """Predict the most common word that follows the given word."""

    if word not in model:
        return None

    counter = Counter(model[word])

    return counter.most_common(1)[0][0]
```

```python
def autocomplete(seed_word, model, length=6):
    """Generate text using the bigram model."""

    words = [seed_word]

    current = seed_word

    for _ in range(length - 1):

        next_word = predict_next_word(current, model)

        if next_word is None:
            break

        words.append(next_word)

        current = next_word

    return " ".join(words)
```

```python
model = build_bigram_model(tokens)
```

```python
print("Starting with 'the'")
print(autocomplete("the", model))

print()

print("Starting with 'cat'")
print(autocomplete("cat", model))

print()

print("Starting with 'dog'")
print(autocomplete("dog", model))
```
